```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/Jane-Pereira/Health-Risks-during-Pregnancy/main/Maternal%20Health%20Risk%20Data%20Set.csv")
```

```python
dataset
```

```python
print(dataset['RiskLevel'].unique())
dataset.info()
```

```python
X = dataset[['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']].values
y = dataset['RiskLevel'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
rcla = RandomForestClassifier(n_estimators=500)
rcla.fit(X_train, y_train)
```

```python
rpred = rcla.predict(X_test)
```

```python
rac = accuracy_score(y_test, rpred)
print("Accuracy: ", rac)
```

```python
print(confusion_matrix(y_test, rpred))
```

```python
def details():
    data = []
    print('Enter patient report details:')

    data.append(int(input('Age: ')))
    data.append(int(input('SystolicBP: ')))
    data.append(int(input('DiastolicBP: ')))
    data.append(float(input('BS: ')))  # Blood Sugar
    data.append(float(input('BodyTemp: ')))
    data.append(int(input('HeartRate: ')))

    return data
```

```python
def prediction(data):
    data = np.array(data)
    data = np.reshape(data, (1, -1))  # Reshape to match the number of features (6 now)
    pred = rcla.predict(data)
    return pred
```

```python
print('- Stage Prediction -')
data = details()

print('Possible Stages: Low, Medium, High')
print('Patient stage: ', prediction(data))
```
